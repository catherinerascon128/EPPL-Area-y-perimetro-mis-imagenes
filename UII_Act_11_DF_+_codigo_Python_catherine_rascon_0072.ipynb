```python
## calcular el area y perimetro de un circulo
# variable A, P, r,
# Formula A = pi*r2 y P = 2 * pi * r
# capturar el valor de r
import math
print("clcular el area y perimetro de un circulo")
r = float(input("ingrese el valor de r: "))
A = math.pi * r**2
P = 2 * math.pi * r
print("el area del circulo es: ", A)
print("el perimetro del circulo es: ", P)
print( " catherine rascon NL = 26")
```
